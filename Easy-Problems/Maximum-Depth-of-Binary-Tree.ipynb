# Maximum Depth of Binary Tree (#104)

The **maximum depth** of a binary tree is the **number of nodes along the longest path from the root node to the farthest leaf node**.

For **LeetCode #104**, depth is measured by **counting nodes**, so:

* An empty tree has a depth of **0**.
* A tree containing only the root has a depth of **1**.

## Key Points

* The path is measured **from the root node down to a leaf node**.
* The answer is the **longest** such path in the tree.
* The reference point is the **root**.
* In LeetCode #104, depth is measured by **counting nodes** (the root contributes **1** to the depth).
* A **leaf node** is a node with **no left child and no right child**.
* `None` is **not** a leaf node. Instead, `None` represents the **absence of a node** (an empty subtree), and its depth is **0**.

## Example

```text
        3
       / \
      9   20
         /  \
        15   7
```

Possible root-to-leaf paths:

* `3 → 9` → Depth = **2**
* `3 → 20 → 15` → Depth = **3**
* `3 → 20 → 7` → Depth = **3**

Therefore, the **maximum depth = 3**.


# Problem statement:
Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

 
## Examples:
### Example 1:
```mermaid

    C["20"]
    D["15"]
    E["7"]

    A -->|Left| B
    A -->|Right| C
    C -->|Left| D
    C -->|Right| E
```
**Input:** root = [3,9,20,null,null,15,7]

**Output:** 3

### Example 2:
```mermaid
graph TD
    A["1 (Root)"]
    B["2"]

    A -->|Right| B
```
**Input:** root = [1,null,2]

**Output:** 2
 

### Constraints:
The number of nodes in the tree is in the range [0, 104].

-100 <= **Node.val** <= 100

# Solution:

````python
from typing import Optional
class TreeNode:
  def __init__(self, val = 0, left = None, right = None):
    self.val = val
    self.left = left
    self.right = right


class Solution:
  def maxDepth(self, root: Optional[TreeNode]) -> int:
    if root is None:
      return 0
    
    return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))
````

# Test Cases:

````python
# Declaring object.

maxdepth = Solution()

root = [3,9,20, None ,None ,15,7]
def build_tree(nodes):
  if not nodes:
    return None
  root_node = TreeNode(nodes[0])
  queue = [root_node]
  i = 1
  while queue and i < len(nodes):
    node = queue.pop(0)
    if nodes[i] is not None:
      node.left = TreeNode(nodes[i])
      queue.append(node.left)
    i += 1
    if i < len(nodes) and nodes[i] is not None:
      node.right = TreeNode(nodes[i])
      queue.append(node.right)
    i += 1
  return root_node

tree_root = build_tree(root)
print("Output:", maxdepth.maxDepth(tree_root))
````
