## Two Sum
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

### Example 1:

**Input:** nums = [2,7,11,15], target = 9 <br>
**Output:** [0,1]<br>
**Explanation:** Because nums[0] + nums[1] == 9, we return [0, 1].

### Example 2:

**Input:** nums = [3,2,4], target = 6 <br>
**Output:** [1,2]

### Example 3:

**Input:** nums = [3,3], target = 6 <br>
**Output:** [0,1] <br>
 
### Constraints:

2 <= nums.length <= 104 <br>
-109 <= nums[i] <= 109 <br>
-109 <= target <= 109 

Only one valid answer exists.

**Follow-up:** Can you come up with an algorithm that is less than O(n2) time complexity?


```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        # Defining empty dictionary
        hashmap = {}

        for i in range(len(nums)):
            complement = target - nums[i]

            # Check complement arrived earlier or not.
            if complement in hashmap:
                # Return the index of the complement (already in hashmap) 
                # and the current index
                return [hashmap[complement], i]

            # Store current value and its index for future lookup.
            hashmap[nums[i]] = i

        """
        for i in range(len(nums)):
          for j in range(i + 1, len(nums)):
            if (nums[i] + nums [j]) == target:
              return [i, j]    
        """
```

## What concepts does Two Sum teach me?
Two Sums teaches three major concepts:
1. **Hash Maps (Dictionary)** 
2. **Complement Lookup**
3. **Trading memory for speed**

The real-world applications come from these concepts, not from literally adding two numbers.
1. Logging system.
2. Banking system.
