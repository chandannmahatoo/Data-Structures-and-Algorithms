# 217. Contains Duplicate

### Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

**Example 1:** <br>
**Input:** nums = [1,2,3,1] <br>
**Output:** true <br>
**Explanation:** <br>
The element 1 occurs at the indices 0 and 3.


**Example 2:** <br>
**Input:** nums = [1,2,3,4] <br>
**Output:** false <br>
**Explanation:** <br>
All elements are distinct.


**Example 3:** <br>
**Input:** nums = [1,1,1,3,3,4,3,2,4,2] <br>
**Output:** true

 
**Constraints:**  <br>
1 <= nums.length <= 105  <br>
-109 <= nums[i] <= 109

```python
from typing import List

class solution:
  def containsDuplicate(self, nums: List[int]) -> bool:

    seen = []

    for i in range(len(nums)):
      if nums[i] in seen:
        return True
      seen.append(nums[i])
    return False
```

## Test Cases:

```python
test_case = solution()

print("List contains Duplicate Value: ",test_case.containsDuplicate([1, 2, 3, 4]))

```

```python
print("List contains Duplicate Value: ",test_case.containsDuplicate([1, 2, 3, 1]))
```

```python
print("List contains Duplicate Value: ",test_case.containsDuplicate([1, 1, 3, 4]))
```

### The Problems
- Since the loop runs n times and each in operation on a list can take up to n comparisons, the worst-case time complexity is: O(n) × O(n) = O(n²)
- The in operator on a list performs a linear search, meaning Python checks each element one by one until it finds the target or reaches the end of the list. This takes O(n) time for each lookup.

I replaced the list with a set because sets are implemented using a hash table. This allows average-case membership checks (in) in O(1) time instead of O(n).

```python
class Solution:
  def containDuplicate(self, nums: List[int]) -> bool:
    seen = set()

    for i in nums:
      if i in seen:
        return True
      
      seen.add(i)
    return False  
```

## Test Cases:

```python
test_case = Solution()

print("List contains Duplicate value: ", test_case.containDuplicate([1,2,3,4,5,6]))
```

```python
print("List contains Duplicate value: ", test_case.containDuplicate([1,2,1,2,4,6]))
```

```python
print("List contains Duplicate value: ", test_case.containDuplicate([1,2,3,4]))
```

### The Most Consice Method:
Converting the list into a set automatically removes duplicate values. If the size of the set is smaller than the size of the original list, duplicates existed.

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        return len(set(nums)) != len(nums)
```

### Test Cases:

```python
print("List contains Duplicate value: ", test_case.containDuplicate([1,2,3,4]))
```

## DSA Takeaway

This problem teaches one of the most important DSA lessons:

* Use a **list** when you need to preserve order or access elements by index.
* Use a **set** when your main operation is **"Have I seen this value before?"** because membership checks are much faster on average.
* Even if two solutions use the same high-level idea ("remember previously seen elements"), the choice of data structure changes the time complexity from **O(n²)** to **O(n)**.

## Comparison Table:
| Approach            | Data Structure | Lookup       | Time  | Space |
| ------------------- | -------------- | ------------ | ----- | ----- |
| Brute Force         | None           | O(n)         | O(n²) | O(1)  |
| Your First Solution | List           | O(n)         | O(n²) | O(n)  |
| Optimized Solution  | Set            | O(1) average | O(n)  | O(n)  |
| One-liner           | Set            | O(1) average | O(n)  | O(n)  |
