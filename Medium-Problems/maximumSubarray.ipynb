# 📝 Problem Statement
Given an integer array nums, find the subarray with the largest sum, and return its sum.

## Example 1:

**Input:** nums = [-2,1,-3,4,-1,2,1,-5,4]<br>
**Output:** 6<br>
**Explanation:** The subarray [4,-1,2,1] has the largest sum 6.


## Example 2:

**Input:** nums = [1]<br>
**Output:** 1<br>
**Explanation:** The subarray [1] has the largest sum 1.


## Example 3:

**Input:** nums = [5,4,-1,7,8]<br>
**Output:** 23<br>
**Explanation:** The subarray [5,4,-1,7,8] has the largest sum 23.
 

## Constraints:

- 1 <= nums.length <= 105<br>
- -104 <= nums[i] <= 104
 

**Follow up:** If you have figured out the O(n) solution, try coding another solution using the divide and conquer approach, which is more subtle.


# 🎯 Understanding the Problem

## What is the problem asking?
- An interger list **nums** is given, the question asked to return **sum** of a **subarray** which is maximum.

Example:
> Suppose **nums = [1 , 3, -1, 3, 5, -6, 7]** is given. After execution of code it must return a list of **Maximum Subarray:  21** whose sum is maximum.


## Brute Force Solution of the Problem:

````python
class Solution:
  def maxSubarray(self, nums):
    ans = float("-inf")
    for i in range(len(nums)):
      total = 0
      for j in range(i, len(nums)):
        total += j
        ans = max(ans, total)

    return ans
````

## Test Cases:

````python
testCase = Solution()

nums = [1 , 3, -1, 3, 5, -6, 7]
print("Maximum Subarray: ", testCase.maxSubarray(nums))
````

````python
nums = [1, 2, 3, 4, -9, -10, 23, -98]
print("Maximum Subarray: ", testCase.maxSubarray(nums))
````

# Optimization of the Problem

## Problem with Brute force code:
It generate every possible subarray but time complexity is O(n²).

## Using Kadane's Algorithm
Suppose you're scanning the array.


### Example:
[-2, 1, -3, 4, -1, 2, 1]
When you reach a number, you have only two choices.


#### Option 1:
Start a brand new subarray here.
> 4
Sum = 4


#### Option 2:
Continue the previous subarray.

Suppose previous best ending here was
> 1
Then
> 1 + 4 = 5

Choose whichever is larger.
So at every index

```python
current_sum =
max(
    nums[i],
    current_sum + nums[i]
)
```
This is **Kadane's Algorithm.**


````python
class Solution:
    def maxSubArray(self, nums):
        current = nums[0]
        best = nums[0]

        for i in range(1, len(nums)):
            current = max(nums[i], current + nums[i])
            best = max(best, current)

        return best
````

## Test Cases:

````python
nums = [1 , 55555, -99999, 3, 5, 10000, -94527]
print("Maximum Subarray: ", testCase.maxSubArray(nums))
````

# Kadane's Algorithm & DP Patterns

## Pattern 1: Contiguous Subarray

If the problem says:

> Find the best contiguous subarray

Immediately think:

- Kadane's Algorithm
- Prefix Sum
- Sliding Window (if constraints fit)

### Examples:
- Maximum Subarray
- Maximum Circular Subarray
- Maximum Product Subarray

## Pattern 2: Extend or Restart

Ask yourself:

> When I reach the current element, do I have two choices?

**Choice 1:**  
Start a new subarray here.

**Choice 2:**  
Continue the previous subarray.

If yes, that's a strong sign of Kadane's Algorithm.

The transition becomes:

```text
current = max(nums[i], current + nums[i])
```

## Pattern 3: DP on Arrays

Define:

```text
dp[i]
```

**Meaning:**  
Best answer ending at index `i`.

For Maximum Subarray:

```text
dp[i] = maximum subarray ending at i
```

**Transition:**

```text
dp[i] = max(nums[i], dp[i-1] + nums[i])
```

If you can define a state like this, you're looking at Dynamic Programming.

## Pattern 4: Local Decision → Global Optimum

Suppose:

```text
current = -50
```

Next number:

```text
20
```

Should we continue?

```text
-50 + 20 = -30
```

or restart?

```text
20
```

Obviously restart.

The previous negative sum can never help future sums.

This greedy local decision is exactly what makes Kadane's Algorithm work.

## Pattern 5: Optimization Problem

Words like:

- Maximum
- Minimum
- Largest
- Best
- Highest

should immediately tell you:

> I'm optimizing something.

Now ask:

> Can I build the answer from previous positions?

If yes,

> think Dynamic Programming.


# 📝 Maximum Subarray (Kadane's Algorithm) – Key Takeaway Notes

## Pattern Recognition

* Contiguous subarray.
* Need **maximum/minimum** sum.
* Optimization problem.
* At each index, choose:

  * Extend previous subarray.
  * Start a new subarray.
* **Think:** Kadane's Algorithm.


## Core Idea

* Maintain the **maximum subarray sum ending at the current index**.
* Previous negative sum is never beneficial.
* Discard negative prefix and restart when necessary.

## DP State

* `dp[i]` = Maximum subarray sum **ending at index `i`**.
* Transition:

  ```python
  dp[i] = max(nums[i], dp[i-1] + nums[i])
  ```
* Kadane optimizes this DP to **O(1)** space.


## Kadane's Formula

```python
current = max(nums[i], current + nums[i])
best = max(best, current)
```


## Initialization

```python
current = nums[0]
best = nums[0]
```

* Never initialize with `0`.
* Handles all-negative arrays correctly.


## Complexity

* **Time:** `O(n)`
* **Space:** `O(1)`


## Mental Checklist

Ask yourself:

* Is it a **contiguous** problem?
* Is it asking for **maximum/minimum**?
* Can I scan the array once?
* At each element, do I decide:

  * Continue?
  * Restart?

If **YES**, think **Kadane's Algorithm**.


## Greedy Insight

* Positive running sum → Keep extending.
* Negative running sum → Restart from current element.
